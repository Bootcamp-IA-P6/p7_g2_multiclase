```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from PIL import Image
from dotenv import load_dotenv
import kagglehub

# Fijar semillas para reproducibilidad
random.seed(42)
np.random.seed(42)

# Cargar credenciales del .env
load_dotenv()
print("Usuario Kaggle:", os.getenv("KAGGLE_USERNAME"))
```

```python
# La primera vez descarga ~8GB — puede tardar varios minutos
# Las siguientes veces es instantáneo (usa la caché)
dataset_path = kagglehub.dataset_download(
    "amerzishminha/forest-fire-smoke-and-non-fire-image-dataset"
)

dataset_path = Path(dataset_path)
print(f"✅ Dataset disponible en:\n{dataset_path}")
print(f"\nContenido:")
for item in sorted(dataset_path.iterdir()):
    print(f"  {item.name}")
```

```python
# Buscar train/ y test/ dentro de la ruta descargada
train_dir = None
test_dir  = None

# Buscar directamente
if (dataset_path / "train").exists():
    train_dir = dataset_path / "train"
    test_dir  = dataset_path / "test"
else:
    # Buscar en subcarpetas
    for subdir in dataset_path.rglob("train"):
        if subdir.is_dir():
            train_dir = subdir
            test_dir  = subdir.parent / "test"
            break

print(f"✅ Train: {train_dir}")
print(f"✅ Test:  {test_dir}")

# Listar clases disponibles
CLASSES = sorted([
    d.name for d in train_dir.iterdir() if d.is_dir()
])
print(f"\nClases detectadas: {CLASSES}")


#Deberías ver:
#✅ Train: /root/.cache/kagglehub/.../train
#✅ Test:  /root/.cache/kagglehub/.../test
#Clases detectadas: ['fire', 'non fire', 'smoke']
```
